```python
import geopandas as gpd
import pandas as pd
import requests
import matplotlib.pyplot as plt
url = "https://gist.githubusercontent.com/meganmason/40a87840b52c5516efc716236a67ccf3/raw/9767072779ac6f22d13e5e4dbf2c7e85f3e6e15e/SNEX20_TS_SP_cogm_swe.geojson"
df = gpd.read_file(url)
df
```

```python
# keep columns we want in the dataframe
df = df[['Site', 'Date/Local Standard Time', 'SWE (mm)']]

df = df.rename({'Date/Local Standard Time': 'date', 'SWE (mm)': 'swe (mm)'}, axis=1)
print(type(df))
```

```python
mo = df.loc[df['Site'] == 'Mesa West Open']
mt = df.loc[df['Site'] == 'Mesa West Trees']
so = df.loc[df['Site'] == 'Skyway Open']
st = df.loc[df['Site'] == 'Skyway Tree']
co = df.loc[df['Site'] == 'County Line Open']
ct = df.loc[df['Site'] == 'County Line Tree']
```

```python

mo.to_csv('../../data/snowpits/mesawest-open_swe.csv', sep=',', index=False, na_rep=' ')
mt.to_csv('../../data/snowpits/mesawest-tree_swe.csv', sep=',', index=False, na_rep=' ')
so.to_csv('../../data/snowpits/skyway-open_swe.csv', sep=',', index=False, na_rep=' ')
st.to_csv('../../data/snowpits/skyway-tree_swe.csv', sep=',', index=False, na_rep=' ')
co.to_csv('../../data/snowpits/countyline-open_swe.csv', sep=',', index=False, na_rep=' ')
ct.to_csv('../../data/snowpits/countyline-tree_swe.csv', sep=',', index=False, na_rep=' ')
```
